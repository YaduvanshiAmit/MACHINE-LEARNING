```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
```

```python
data.head()
```

```python
# Univariate Analysis
# first we see hw many category is there
data['species'].value_counts()
```

```python
# then we makeeach category different data
data_setosa = data.loc[data['species']=='setosa']
data_virginica = data.loc[data['species']=='virginica']
data_versicolor = data.loc[data['species']=='versicolor']
```

```python
# nw we plot
plt.plot(data_setosa['sepal_length'],np.zeros_like(data_setosa['sepal_length']),'o')
plt.plot(data_virginica['sepal_length'],np.zeros_like(data_virginica['sepal_length']),'_')
plt.plot(data_versicolor['sepal_length'],np.zeros_like(data_versicolor['sepal_length']),'+')
plt.xlabel('sepal legth')
plt.show()
```

Doing univariate analysis using one feature sepal_legth we see that poits are not clearly separated 
Lets see other feature also

```python
plt.plot(data_setosa['sepal_width'],np.zeros_like(data_setosa['sepal_width']),'o')
plt.plot(data_virginica['sepal_width'],np.zeros_like(data_virginica['sepal_width']),'-')
plt.plot(data_versicolor['sepal_width'],np.zeros_like(data_versicolor['sepal_width']),'+')
plt.xlabel('sepal width')
plt.show()
```

```python
plt.plot(data_setosa['petal_length'],np.zeros_like(data_setosa['petal_length']),'o')
plt.plot(data_virginica['petal_length'],np.zeros_like(data_virginica['petal_length']),'-')
plt.plot(data_versicolor['petal_length'],np.zeros_like(data_versicolor['petal_length']),'+')
plt.xlabel('petal legth')
plt.show()
```

here we clearly see one category simple separate but other two are not petal legth of setosa are less than 3 

```python
plt.plot(data_setosa['petal_width'],np.zeros_like(data_setosa['petal_width']),'o')
plt.plot(data_virginica['petal_width'],np.zeros_like(data_virginica['petal_width']),'o')
plt.plot(data_versicolor['petal_width'],np.zeros_like(data_versicolor['petal_width']),'+')
plt.xlabel(' petal_width')
plt.show()
```

From the univariate analysis we know that by using feature petal length and petal width we easily separate sentosa and virginica or versicolor
